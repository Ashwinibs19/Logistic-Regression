```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#Load the data set
bank = pd.read_csv("bank-full.csv",sep=";")
bank
```

# EDA

```python
bank.info()
```

```python
bank
```

```python
# One-Hot Encoding of categrical variables
data1=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
# Custom Binary Encoding of Binary o/p variables 
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

```python
#Shape of the dataset
data1.shape
```

```python
# Removing NA values in data set
data1 = data1.dropna()
data1.shape
```

# Model Building

```python
# Dividing our data into input and output variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

# Model Predictions

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

# Testing Model Accuracy

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y,y_pred)
#confusion_matrix
print('Confusion matrix\n\n', cm)

TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]

print('\nTrue Positives(TP) = ', TP)

print('\nTrue Negatives(TN) = ', TN)

print('\nFalse Positives(FP) = ', FP)

print('\nFalse Negatives(FN) = ', FN)
```

```python
The confusion matrix shows 39183 + 739 = 39922 correct predictions and 1129 + 4160 = 5289 incorrect predictions.

In this case, we have

True Positives (Actual Positive:1 and Predict Positive:1) - 39922
True Negatives (Actual Negative:0 and Predict Negative:0) - 5289
False Positives (Actual Negative:0 but Predict Positive:1) - 739 (Type I error) 
False Negatives (Actual Positive:1 but Predict Negative:0) - 4160 (Type II error)
```

```python
# visualize confusion matrix with seaborn heatmap
import seaborn as sns

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39183+1129)/(39183+739+4160+1129)
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
